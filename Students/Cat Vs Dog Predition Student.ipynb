```python
import tensorflow.keras.backend as k
if k.image_data_format()=='channels_first':
  input_chape=(3,img_width,img_height)
else:
  input_shape=(img_width,img_height,3)
```

**Image generation to train the model**

```python
train_datagen=ImageDataGenerator(rescale=None,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen=ImageDataGenerator (rescale=None)              # bringing rgb value from0-255 to range of 0-1
train_generator=train_datagen.flow_from_directory(train_data_dir,target_size=(img_width,img_height), batch_size=None , class_mode=None)#,classes=['cats','dogs'])
validation_generator=test_datagen.flow_from_directory(validation_data_dir,target_size=(img_width,img_height),batch_size=None,class_mode=None)
```

```python
model=Sequential()
model.add(Conv2D(64,(3,3),input_shape=input_shape))   #64 neurons using 3X3 filter

model.add(Activation('None'))
model.add(MaxPooling2D(pool_size=(2,2)))#pooling reduce the size of data

model.add(Flatten())#conversion of multidimensional arrar to 1D array
model.add(Dense(64))

model.add(Activation('None'))
model.add(Dense(1))#output layer
model.add(Activation('None'))
model.summary()
```

Model compiling

```python
model.compile(optimizer="None",loss="None",metrics=['None'])
model.summary()
```

model fitting

```python
training = model.fit(train_generator,steps_per_epoch=nb_train_sample,epochs=epochs,validation_data=validation_generator,validation_steps=nb_validation_samples,)
```
